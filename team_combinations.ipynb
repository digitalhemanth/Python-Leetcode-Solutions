```python
from itertools import combinations

# List of IPL teams
teams = [
    "Chennai Super Kings (CSK)", 
    "Mumbai Indians (MI)", 
    "Royal Challengers Bangalore (RCB)", 
    "Delhi Capitals (DC)", 
    "Sunrisers Hyderabad (SRH)", 
    "Kolkata Knight Riders (KKR)", 
    "Rajasthan Royals (RR)", 
    "Punjab Kings (PBKS)", 
    "Lucknow Super Giants (LSG)", 
    "Gujarat Titans (GT)"
]

# Generate all possible combinations of matches between two teams
matches = list(combinations(teams, 2))

# Display the matches
print(f"Total possible matches: {len(matches)}\n")
for match in matches:
    print(f"{match[0]} vs {match[1]}")
```

```python
# List of teams
teams = ["CSK", "MI", "RCB", "DC", "SRH", "KKR", "RR", "PBKS", "LSG", "GT"]

# Brute force approach
matches = []
for i in range(len(teams)):
    for j in range(i + 1, len(teams)):  # Ensure no repetition
        matches.append((teams[i], teams[j]))

# Display the matches
print(f"Total possible matches: {len(matches)}\n")
for match in matches:
    print(f"{match[0]} vs {match[1]}")
```
